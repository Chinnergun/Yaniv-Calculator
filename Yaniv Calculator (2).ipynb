```python
# Test cell 001
def yaniv_play():
        
    """
    DOCSTRING:
    This is a function for calculating yaniv scores 
    
    variables to determine:
    1. Number of PLAYERS who are playing
    2. Score LIMIT
    3. Score DEDUCTION checkpoints (and deduction quantities)
    4. Whether we END the game the moment at least 1 player busts
    5.
    
    
    Created by: Ong Chin Kang
    """
    
    # Phase 1: Preparation
    # Part I: no. of players, who they are & score limit
    while True:
        try:
            # introductory greeting
            print("\nLet's play Yaniv!")

            # how many players
            no_of_players = int(input("\n How many of you are playing? \n")) # accept only integers

            # create new player profiles
            players = {} # store the player profiles
            for count in list(range(0, no_of_players)):
                player_name = input("Please enter your name: ")
                players[player_name] = 0

            # Welcome message for all players
            print("\nWelcome to a new game of Yaniv, ")
            for name,score in players.items():
                print(f"{name}")

            # set score limit
            score_limit = int(input("\nWhat is the score limit? \n")) # accept only integers
        except:
            print("Eh, why didn't you enter a whole number ah? \nSee lah, now we have to restart the process!")
        else:
            break # concludes part I of the prepararatory process
    
    settingbustreset = True
    while settingbustreset:
        # part II of phase 1: bust reset

            # Decide if gameplay continues even after game busts
        bust_reset = input("\nEnd game after someone busts (Y/N)? \n").upper() # accept only Y or N
        if bust_reset == 'Y' or bust_reset == "N":
            settingbustreset = False # conclude phase 1b
        else:
            print("\nError: Please input 'Y' or 'N' ")
            
    # Part III of phase 1: set the deduction checkpoints
    deduct_bonus = {} # this dictionary will store checkpoints and respective deductions
    setting_checkpoints = True # enables players to continue setting until otherwise stated

    while setting_checkpoints:
        try:
            # add new checkpoint & deduction value first
            new_checkpoint = int(input("\nPlease enter a new checkpoint value: \n")) # expect an integer
            new_deduction = int(input("\nPlease enter a value to deduct at the given checkpoint: \n")) # expect an integer
            deduct_bonus[new_checkpoint] = new_deduction # add the new pair into dictionary

            deciding = True # will player add new checkpoint?
            while deciding:
                cont_add_chkpts = input("\nWould you like to add a new deduction checkpoint? (Y/N) \n").upper()
                if cont_add_chkpts == 'N':
                    deciding = False # a decision has been made
                    setting_checkpoints = False # You are done setting checkpoints and deductions. Break out of while loop
                elif cont_add_chkpts == 'Y':
                    deciding = False # a decision has been made
                    pass # players can continue adding
                else:
                    print("\n Error: Please enter either 'Y' or 'N' \n") # continue to decide
        except:
            print("\nError: Please input an integer. \n")


    print("\nHere is a review of the checkpoints & their respective deductible bonuses: ") # provide an overview
    for chkpt,deduct in deduct_bonus.items():
        print(f"At {chkpt} points, {deduct} points will be deducted from a given player's score.")
                

    # commence game greeting    
    print("\nGame setup complete. Let's begin!")
    print("-"*100)
            
    # Phase 2: gameplay
    playing = True # game runs until otherwise indicated
    outboard = {} # dictionary of the players who are already out of the game, in case we DO NOT reset after someone busts
    winner = [] # empty list object for the winner if scenario is DO NOT END after 1st bust
    while playing:
        try:
            # indicate the beginning of a new round
            print("\nA new round has begun.\n")
                
            # update the scores of the players for a given round
            for name,score in players.items():

                # record the round score of a given player
                round_score = int(input(f"{name}'s round score: ")) # expect an integer

                # add the player's score
                players[name] += round_score
                print(f"{name}: {players[name]}") # display a player's updated score
                
            # check if any of the players hit a checkpoint
            for name,score in players.items():
                for chkpt,deduct in deduct_bonus.items():
                    if score == chkpt: # if someone's score hits a checkpoint
                        players[name] -= deduct # bonus deduction will apply
                    
                
        except: # if someone keys something wrongly or tries something funny
            print("Oi, mai kee xiao. \nPls enter a whole number.")
            
        else: # should there be no error
            # step 1: do limit checks
            for name,score in players.items():
                if score > score_limit:
                    if bust_reset == 'Y': # if we end the game the moment someone busts
                        # end the game
                        print("\nGame Over") # indicate that the game is over
                        playing = False # cut the game
                        outboard[name] = score # Update to show the player who made the bust
                    else: # if we play til last man standing
                        outboard[name] = score # update outboard together with bust score
           
            # remaining code is for the case when the game continues until 1 person is left
            # expel players who alr bust
            for name,score in outboard.items():
                if name in players.keys(): # if a player who is in the outboard is still part of the players' profiles
                    players.pop(name) # remove player from players' profile
                    
            # Display Updated outboard
            print("\nOutboard:")
            for name,score in outboard.items():
                print(f"{name}: {score}")
            if len(outboard) == 0:
                print("NA")
            
            # Display Updated scoreboard
            print("\nScoreboard")
            for name,score in players.items():
                print(f"{name}: {score}")
            print("-"*100)
            

            
            # check to see if there is only 1 player left 
            if len(players) == 1:
                winner = list(players.keys())[0]
                print(f"Congratulations, {winner} has won the game!")
                playing = False # end the game
            
            
```
